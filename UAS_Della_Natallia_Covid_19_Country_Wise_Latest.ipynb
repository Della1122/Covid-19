DROP
New cases, New deaths, New recovered, Deaths / 100 Cases, Recovered / 100 Cases, Deaths / 100 Recovered, Confirmed last week, 1 week change, 1 week % increase

Alasan
1. New cases, New deaths, New recovered: Kolom-kolom ini tidak diperlukan karena informasi tentang kasus baru, kematian baru, dan pemulihan baru sudah diwakili oleh kolom lain seperti Confirmed, Deaths, dan Recovered. Selain itu, fokus pada kolom-kolom utama ini akan lebih mudah dipahami.

2. Deaths / 100 Cases, Recovered / 100 Cases, Deaths / 100 Recovered: Kolom-kolom ini menyajikan angka dalam bentuk persentase yang dapat dihitung dari kolom lainnya (Deaths, Recovered, dan Confirmed). Oleh karena itu, kolom ini mungkin tidak perlu disimpan terpisah.

3. Confirmed last week, 1 week change, 1 week % increase: Kolom-kolom ini memberikan informasi tentang perubahan jumlah kasus dalam satu minggu terakhir. Namun, jika fokus analisis saat ini adalah pada data harian, kolom-kolom ini mungkin tidak relevan.

```python
# Menghapus kolom yang tidak diperlukan (columns='New cases', 'New deaths', 'New recovered', 'Deaths / 100 Cases', 'Recovered / 100 Cases', 'Deaths / 100 Recovered', 'Confirmed last week', '1 week change', '1 week % increase')
cols_to_drop = ['New cases', 'New deaths', 'New recovered', 'Deaths / 100 Cases', 'Recovered / 100 Cases', 'Deaths / 100 Recovered', 'Confirmed last week', '1 week change', '1 week % increase']
df.drop(columns=cols_to_drop, inplace=True)

# Menampilkan 5 baris pertama dari DataFrame
df.head()
```
